**Question 1:**
You are building a landing page for a small bakery in your city. The page looks fine on your laptop but when you open it on your phone, the text is too small and the images go outside the screen. Which CSS technique will you use to fix this and how will you start?

**Answer:**
To solve this problem, I would use Responsive Web Design with CSS Media Queries. Responsive design allows a website to adjust its layout according to different screen sizes, making it easy to use on desktops, tablets, and mobile devices.

First, I would add the viewport meta tag in the HTML file:

<meta name="viewport" content="width=device-width, initial-scale=1.0">

Then, I would use CSS media queries to adjust the font size, image size, and layout for smaller screens. I would also set the image width to 100% so that images do not overflow the screen.

Example:

img {
max-width: 100%;
height: auto;
}

@media (max-width: 600px) {
body {
font-size: 16px;
}
}

This makes the webpage responsive and easy to read on mobile devices.

**Question 2:**
You added a navbar at the top of your website. The client wants the navbar to always stay visible at the top, even when the user scrolls down. Which CSS property will you use and what value will you set?

**Answer:**
I would use the position property with the value fixed.

Example:
.navbar {
position: fixed;
top: 0;
width: 100%;
}

The position: fixed property keeps the navbar fixed at the top of the browser window even when the user scrolls. The top: 0 value places it at the top, and width: 100% ensures it covers the full width of the page.

**Question 3:**
You wrote CSS for a button — first you added a class .btn with blue background, then added an id #submitBtn with red background to the same button. The button is showing red instead of blue. Why is this happening?

**Answer:**
The button appears red because an ID selector has higher specificity than a class selector in CSS.

Example:

.btn {
background-color: blue;
}

#submitBtn{
background-color: red;
}

Even though the class is written first, the ID selector has higher priority, so the button displays a red background.

**Question 4:**
You are building a product card for an online store. You want three cards to sit side by side on desktop, but stack one below the other on mobile. Which layout system will you use and how will you make it responsive?

**Answer:**
I would use CSS Flexbox because it is simple and effective for creating responsive layouts.

Example:
.container {
display: flex;
gap: 20px;
}

.card {
flex: 1;
}

@media (max-width: 768px) {
.container {
flex-direction: column;
}
}

On desktop screens, the cards are displayed side by side. On mobile screens, the media query changes the layout to a column, so the cards are displayed one below another.

**Question 5:**
You have a section with text and you want to add some space inside the section between the text and the border, and also some space outside the section to separate it from the next section. Which two CSS properties will you use and how will they differ?

**Answer:**
I would use the padding and margin properties.

Padding adds space inside the element, between the content and the border.

Margin adds space outside the element, creating distance between the current element and other elements.

Example:
.section {
padding: 20px;
margin-bottom: 30px;
}

Padding improves the spacing inside the section, while margin separates it from the next section.

**Question 6:**
Your senior gives you a design where a 'SALE' label has to sit on the top-right corner of every product image. The label should overlap the image. Which positioning approach will you use?

**Answer:**
I would use position: relative on the product container and position: absolute on the SALE label.

Example:
.product {
position: relative;
}

.sale {
position: absolute;
top: 10px;
right: 10px;
}

This allows the SALE label to overlap the image and stay fixed at the top-right corner of each product image.

**Question 7:**
What is the difference between display: none and visibility: hidden? Give one situation where you would use each.

**Answer:**
display: none completely removes the element from the webpage layout, and it does not occupy any space.

visibility: hidden hides the element but still reserves its original space in the layout.

Example:
.box1 {
display: none;
}

.box2 {
visibility: hidden;
}

I would use display: none for menus, popups, or elements that should disappear completely.

I would use visibility: hidden when I want to temporarily hide an element without affecting the page layout.

**Question 8:**
What is the difference between Flexbox and CSS Grid? When should you use each?

**Answer:**
Flexbox is a one-dimensional layout system. It arranges items in either a row or a column and is ideal for navigation bars, buttons, and simple layouts.

CSS Grid is a two-dimensional layout system. It controls both rows and columns, making it suitable for complex webpage layouts.

I would use Flexbox for simple component layouts and CSS Grid for complete webpage layouts with multiple rows and columns.

**Question 9:**
What are the three ways to add CSS to an HTML page? Which one is preferred for real projects and why?

**Answer:**
There are three ways to add CSS to an HTML page:

Inline CSS – CSS is written directly inside an HTML element using the style attribute.

Example:
<p style="color: blue;">Hello</p>

Internal CSS – CSS is written inside the <style> tag in the HTML document.

Example:

<style> p { color: blue; } </style>

External CSS – CSS is written in a separate .css file and linked to the HTML page.

Example:
<link rel="stylesheet" href="style.css">

The preferred method for real projects is External CSS because it keeps the HTML and CSS separate, improves code readability, makes maintenance easier, and allows the same stylesheet to be reused across multiple pages.

**Question 10:**
What is a media query in CSS? Write a small example that changes the background color when the screen is smaller than 600px.

**Answer:**
A media query is a CSS feature that allows different styles to be applied based on the screen size or device characteristics. It is mainly used to make websites responsive.

Example:

body {
background-color: white;
}

@media (max-width: 600px) {
body {
background-color: lightblue;
}
}

In this example, the background color changes from white to light blue when the screen width becomes 600 pixels or smaller.
